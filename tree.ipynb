# Trees

## binary tree

```python
class Node:
    def __init__(self, val, left = None, right = None):
        self.left = left
        self.right = right
        self.val = val
        
    def preorder(self):
        print(self)
        if self.left: self.left.preorder()
        if self.right: self.right.preorder()

    def inorder(self):
        if self.left: self.left.inorder()
        print(self)
        if self.right: self.right.inorder()

    def postorder(self):
        if self.left: self.left.postorder()
        if self.right: self.right.postorder()
        print(self)

    def __repr__(self):
        return self.val

class Tree:
    def __init__(self, head):
        self.head = head

    def DFS(self):
        self.head.postorder()
        
    def BFS(self):
        self.head.preorder()

```

```python
root = Node("A")
t1 = Tree(root)
t1.head.left = Node("B")
t1.head.right = Node("C")
t1.head.left.left = Node("D")
t1.head.left.right = Node("E")
t1.head.right.left = Node("F")

t1.BFS()
print()
t1.DFS()
print()
t1.head.inorder()
```

# Binary search tree

```python
class Node:
    def __init__(self, val, left = None, right = None):
        self.left = left
        self.right = right
        self.val = val
        
    def preorder(self):
        print(self.val)
        if self.left: self.left.preorder()
        if self.right: self.right.preorder()

    def inorder(self):
        if self.left: self.left.inorder()
        print(self.val)
        if self.right: self.right.inorder()

    def postorder(self):
        if self.left: self.left.postorder()
        if self.right: self.right.postorder()
        print(self.val)                
        
    def append(self, val):
        if self.left and val<self.val:
            self.left.append(val)
        elif val<self.val:
            self.left = Node(val)
        if self.right and val>self.val:
            self.right.append(val)
        elif val>self.val:
            self.right = Node(val)
    
    def find_min(self):
        current = self
        while current.left:
            current = current.left
        return current
    
    def delete(self, val, parent=None):
        if val < self.val:
            if self.left:
                return self.left.delete(val, self)
            return False 
        elif val > self.val:
            if self.right:
                return self.right.delete(val, self)
            return False
        else:
            if self.left and self.right:
                successor = self.right.find_min()
                self.val = successor.val
                self.right.delete(successor.val, self)
            elif parent:
                if parent.left == self:
                    parent.left = self.left if self.left else self.right
                else:
                    parent.right = self.left if self.left else self.right
            else:
                if self.left:
                    self.val = self.left.val
                    self.right = self.left.right
                    self.left = self.left.left
                elif self.right:
                    self.val = self.right.val
                    self.left = self.right.left
                    self.right = self.right.right
                else:
                    self.val = None
            return True
    
    def _repr_helper(self, level=0):
        indent = "    " * level
        out = f"{indent}{self.val}\n"
        if self.left:
            out += f"{indent}  L: " + self.left._repr_helper(level + 1).lstrip()
        if self.right:
            out += f"{indent}  R: " + self.right._repr_helper(level + 1).lstrip()
        return out
            
    def __repr__(self):
        return self._repr_helper()


class Tree:
    def __init__(self, head):
        self.head = head

    def __repr__(self):
        return repr(self.head)
    def DFS(self):
        self.head.postorder()
        
    def BFS(self):
        self.head.preorder()

    def append(self,val):
        self.head.append(val)

    def remove(self, val):
        if self.head is None:
            return False
        return self.head.delete(val)
```

```python
root = Node(100)
t1 = Tree(root)
t1.append(400)
t1.append(60)
t1.append(70)
t1.append(30)
t1.append(31)
t1.append(29)

t1.DFS()
print()
t1.BFS()
print()
t1.head.inorder()
print()
print(t1.head)
t1.remove(30)
print(t1.head)
```
